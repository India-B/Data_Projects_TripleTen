```python
# =========================
# 4) MATCHABILITY CHECK (do we have a clean join key across files?)
# =========================

# Standardize city names to compare overlap
demo_cities  = set(demo["City"].astype(str).str.strip().str.lower().unique())
train_cities = set(train["City"].astype(str).str.strip().str.lower().unique())
overlap = demo_cities.intersection(train_cities)

print("Overlap between demographics cities and train cities:", len(overlap))
print("Overlap examples:", sorted(list(overlap))[:25])
```

```python
# drop junk index column if present
ua_clean = ua.copy()
if "Unnamed: 0" in ua_clean.columns:
    ua_clean = ua_clean.drop(columns=["Unnamed: 0"])

# list of US states for filtering (because UA_Country holds state names for US entries)
US_STATES = {
    "Alabama","Alaska","Arizona","Arkansas","California","Colorado","Connecticut","Delaware",
    "Florida","Georgia","Hawaii","Idaho","Illinois","Indiana","Iowa","Kansas","Kentucky","Louisiana",
    "Maine","Maryland","Massachusetts","Michigan","Minnesota","Mississippi","Missouri","Montana",
    "Nebraska","Nevada","New Hampshire","New Jersey","New Mexico","New York","North Carolina",
    "North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania","Rhode Island","South Carolina",
    "South Dakota","Tennessee","Texas","Utah","Vermont","Virginia","Washington","West Virginia",
    "Wisconsin","Wyoming","District of Columbia"
}

# Primary filter: North America + "UA_Country" is a US state name
ua_us = ua_clean[
    (ua_clean["UA_Continent"] == "North America") &
    (ua_clean["UA_Country"].isin(US_STATES))
].copy()

print("uaScores total rows:", ua_clean.shape[0])
print("uaScores US rows (state-based filter):", ua_us.shape[0])

# sanity peek: what 'countries' exist in North America?
print("\nTop UA_Country values in North America (first 25):")
display(
    ua_clean.loc[ua_clean["UA_Continent"]=="North America", "UA_Country"]
    .value_counts().head(25)
)

display(ua_us[["UA_Name","UA_Country","UA_Continent"]].head(10))
```

```python
# =========================
# STEP 2B) MAKE us_demographics CITY-LEVEL (pivot Race -> columns)
# =========================

demo_clean = demo.copy()

# 1) City-level numeric table (mean works here because values repeat across Race rows)
demo_numeric_cols = [
    "Median Age","Male Population","Female Population","Total Population",
    "Number of Veterans","Foreign-born","Average Household Size"
]

demo_city = (
    demo_clean
    .groupby(["City","State"], as_index=False)[demo_numeric_cols]
    .mean()
)

# 2) Pivot race counts wide: one row per City+State
race_wide = (
    demo_clean
    .pivot_table(index=["City","State"], columns="Race", values="Count", aggfunc="sum")
    .reset_index()
)

# 3) Merge numeric + race
demo_city = demo_city.merge(race_wide, on=["City","State"], how="left")

# 4) Create race shares (optional but useful)
race_cols = [c for c in race_wide.columns if c not in ["City","State"]]
for c in race_cols:
    demo_city[c + "_share"] = demo_city[c] / demo_city["Total Population"]

print("demo original rows:", demo.shape)
print("demo city-level rows:", demo_city.shape)
display(demo_city.head(5))
```

```python
# =========================
# STEP 2C) CREATE SIMPLE JOIN KEYS + TRY MERGE (UA US) <-> (Demographics)
# =========================


def norm_text(s):
    """Beginner-friendly normalization: lowercase, strip, remove non-letters/numbers."""
    s = str(s).strip().lower()
    s = re.sub(r"[^a-z0-9]+", "", s)
    return s

ua_us["city_key"] = ua_us["UA_Name"].apply(norm_text)
ua_us["state_key"] = ua_us["UA_Country"].apply(norm_text)

demo_city["city_key"] = demo_city["City"].apply(norm_text)
demo_city["state_key"] = demo_city["State"].apply(norm_text)

merged_us = ua_us.merge(
    demo_city,
    on=["city_key","state_key"],
    how="left",
    suffixes=("_ua","_demo")
)

print("UA US rows:", ua_us.shape[0])
print("Merged rows:", merged_us.shape[0])

# How many got demographics?
matched = merged_us["Total Population"].notna().sum()
print("Matched demographics rows:", matched)
print("Match rate:", round(matched / merged_us.shape[0] * 100, 1), "%")

# Show a few unmatched examples (to see naming issues)
unmatched_examples = merged_us.loc[merged_us["Total Population"].isna(), ["UA_Name","UA_Country"]].head(20)
print("\nUnmatched UA cities (examples):")
display(unmatched_examples)

# Save the final working table for next steps
display(merged_us.head(5))
```

### (Clustering prep only) — clean features + pick number of clusters
Goal: produce a clean matrix for clustering and choose k (how many clusters) using simple metrics.

### Pick features + drop broken/odd stuff
We’re clustering on UA livability dimensions (not demographics yet).

```python
# Core UA features for clustering
cluster_features = [
    "Housing",
    "Cost of Living",
    "Safety",
    "Healthcare",
    "Education",
    "Environmental Quality",
    "Commute",
    "Startups",
    "Venture Capital"
]

# Economy is excluded because it looks constant / unusable in our plots & heatmap
# (we can confirm later with nunique, but for now keep it out)

df_clust = df.copy()

# Keep only the columns we need + labels for later
keep_cols = ["UA_Name", "UA_Country_clean"] + cluster_features
df_clust = df_clust[keep_cols].copy()

# Check for weird values (like Cost of Living near 0)
print("Min values:")
display(df_clust[cluster_features].min().to_frame("min"))

print("\nMax values:")
display(df_clust[cluster_features].max().to_frame("max"))

# Find suspicious Cost of Living rows
sus = df_clust[df_clust["Cost of Living"] < 1][["UA_Name","UA_Country_clean","Cost of Living"]]
print("\nSuspicious Cost of Living (<1) rows:")
display(sus)
```

### Decide what to do with the outlier (beginner option: remove it)
If you see 1 row with Cost of Living ~0, remove it for clustering stability.

```python
# Remove any rows with Cost of Living < 1 (simple outlier rule)
df_clust = df_clust[df_clust["Cost of Living"] >= 1].copy()

print("Rows after removing COL<1:", df_clust.shape[0])
```

### Standardize (very important for clustering)
Clustering needs features on similar scales.

```python
X = df_clust[cluster_features].copy()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Scaled matrix shape:", X_scaled.shape)
```

### Choose k using silhouette score (simple + defensible)

```python
scores = []

# We'll test k from 2 to 8 (small dataset: 60ish cities)
for k in range(2, 9):
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = model.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    scores.append((k, score))

scores_df = pd.DataFrame(scores, columns=["k", "silhouette"])
display(scores_df)

best_k = scores_df.sort_values("silhouette", ascending=False).iloc[0]["k"]
print("Best k by silhouette:", int(best_k))
```

#### What we learned:
6 cities have Cost of Living = 0 (Bozeman, Fort Collins, Palo Alto, Portland (ME), Providence, San Luis Obispo). That’s almost certainly a data coding issue, not “free living,” so removing them for clustering is reasonable.

Silhouette says k=2 is best (0.257). That means the data most strongly supports two broad livability profiles in this US subset.

### (Modeling) — fit KMeans (k=2) + interpret profiles + 2 visualizations
Goal: create the actual “livability profiles,” name them in plain English, and generate visuals that can be used in both notebook + policy brief.

#### Fit final KMeans and attach cluster labels:

```python
k = 2  # best by silhouette
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_scaled)

df_clust["cluster"] = labels

print(df_clust["cluster"].value_counts())
display(df_clust[["UA_Name", "UA_Country_clean", "cluster"]].head(10))
```

#### Profile each cluster (means table):
This is where the “story” comes from.

```python
cluster_means = df_clust.groupby("cluster")[cluster_features].mean().round(2)
display(cluster_means)

# Optional: show differences (cluster 1 - cluster 0) to see what separates them
diff = (cluster_means.loc[1] - cluster_means.loc[0]).sort_values(ascending=False)
print("Cluster 1 minus Cluster 0 (positive means cluster 1 higher):")
display(diff.to_frame("difference"))
```

#### Visualization 1: PCA scatter (cities colored by profile):

```python
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

plt.figure()
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df_clust["cluster"])
plt.xlabel("PCA component 1")
plt.ylabel("PCA component 2")
plt.title("US Cities: Livability profiles (KMeans clusters) in 2D (PCA)")
plt.grid(True)
plt.show()

print("Explained variance by PCA components:", pca.explained_variance_ratio_.round(3))
```

#### Visualization 2: Cluster averages (bar chart):
Simple, interpretable, policy-friendly.

```python
# Plot mean scores per feature for each cluster
cluster_means_T = cluster_means.T

plt.figure()
for cl in cluster_means_T.columns:
    plt.plot(cluster_means_T.index, cluster_means_T[cl], marker="o", label=f"Cluster {cl}")

plt.xticks(rotation=90)
plt.ylabel("Average UA score")
plt.title("Cluster feature averages (livability profiles)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

#### Quick “city examples” from each profile:
This helps you write the narrative fast.

```python
# choose a "representative" city: closest to the cluster center in scaled space

centers = kmeans.cluster_centers_
X_scaled_arr = np.array(X_scaled)

rep_rows = []
for cl in range(k):
    idxs = np.where(df_clust["cluster"].values == cl)[0]
    dists = np.linalg.norm(X_scaled_arr[idxs] - centers[cl], axis=1)
    rep_idx = idxs[np.argmin(dists)]
    rep_rows.append(df_clust.iloc[rep_idx][["UA_Name","UA_Country_clean","cluster"]])

rep_df = pd.DataFrame(rep_rows)
print("Representative cities (closest to cluster centers):")
display(rep_df)

# show a few cities per cluster
for cl in range(k):
    print(f"\nCluster {cl} examples:")
    display(df_clust[df_clust["cluster"] == cl][["UA_Name","UA_Country_clean"]].head(10))
```
